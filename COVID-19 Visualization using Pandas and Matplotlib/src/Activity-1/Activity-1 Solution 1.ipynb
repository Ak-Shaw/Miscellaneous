```python
#import libraries
import pandas as pd # Used for handling the dataset
import matplotlib.pyplot as plt # Used for plottig the analysed dataset

# Setting Layout
%matplotlib inline # To include the plot(frontend) just besides the code(backend)
plt.style.use('fivethirtyeight') # Setting a beautiful style for our plots
```

```python
#Mini Task A: Read a file from url
data = pd.read_csv('https://raw.githubusercontent.com/datasets/covid-19/master/data/countries-aggregated.csv')

# print top 5 rows
data.head()
```

```python
#Mini Task B: Make a new column Total Cases
data['Total Cases'] = data.Confirmed + data.Recovered + data.Deaths

# print top 5 rows of new data
data.head()
```

# Optional

To make values look more visually appealing and to easily analyse the total cases, we can use **background_gradient** utility provided by pandas as shown below.

```python
# calculate total cases per country till date
x = data.groupby('Country')['Total Cases'].sum().to_frame()

# apply styling
x.style.background_gradient('RdGy')
```
